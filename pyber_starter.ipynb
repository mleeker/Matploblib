```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
merged_data = pd.merge(city_data, ride_data, on="city")
#merged_data["fare"] = merged_data["fare"].map("${:.2f}".format)

# Display the data table for preview
merged_data.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
#avg fare

bub_urban_fare = merged_data.loc[merged_data["type"] == "Urban", :]
urban_city_fare = bub_urban_fare.groupby(["city"])
bub_urban_fare_avg = urban_city_fare["fare"].mean()


bub_rural_fare = merged_data.loc[merged_data["type"] == "Rural", :]
rural_city_fare = bub_rural_fare.groupby(["city"])
bub_rural_fare_avg = rural_city_fare["fare"].mean()

bub_sub_fare = merged_data.loc[merged_data["type"] == "Suburban", :]
sub_city_fare = bub_sub_fare.groupby(["city"])
bub_sub_fare_avg = sub_city_fare["fare"].mean()


#total # of riders
bub_urban_riders = merged_data.loc[merged_data["type"] == "Urban", :]
urban_city_riders = bub_urban_riders.groupby(["city"])
bub_urban_riders_sum = urban_city_riders["ride_id"].count()

bub_rural_riders = merged_data.loc[merged_data["type"] == "Rural", :]
rural_city_riders = bub_rural_riders.groupby(["city"])
bub_rural_riders_sum = rural_city_riders["ride_id"].count()

bub_sub_riders = merged_data.loc[merged_data["type"] == "Suburban", :]
sub_city_riders = bub_sub_riders.groupby(["city"])
bub_sub_riders_sum = sub_city_riders["ride_id"].count()

#driver count
bub_urban_driver = merged_data.loc[merged_data["type"] == "Urban", :]
urban_city_driver = bub_urban_driver.groupby(["city"])
bub_urban_driver_count = urban_city_driver["driver_count"].sum()

bub_rural_driver = merged_data.loc[merged_data["type"] == "Rural", :]
rural_city_driver = bub_rural_driver.groupby(["city"])
bub_rural_driver_count = rural_city_driver["driver_count"].sum()

bub_sub_driver = merged_data.loc[merged_data["type"] == "Suburban", :]
sub_city_driver = bub_sub_driver.groupby(["city"])
bub_sub_driver_count = sub_city_driver["driver_count"].sum()

# Build the scatter plots for each city types
#urban
urban = plt.scatter(bub_urban_riders_sum, bub_urban_fare_avg, marker="o", facecolors="gold", edgecolors="goldenrod",
            s=bub_urban_driver_count, alpha=.8, label="Urban")

rural = plt.scatter(bub_rural_riders_sum, bub_rural_fare_avg, marker="o", facecolors="lightcoral", edgecolors="indianred",
            s=bub_rural_driver_count, alpha=.8, label="Rural")

suburban = plt.scatter(bub_sub_riders_sum, bub_sub_fare_avg, marker="o", facecolors="lightskyblue", edgecolors="dodgerblue",
            s=bub_sub_driver_count, alpha=.8, label="Suburban")

# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data (2016)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid()

# Create a legend
marker_size = 1
plt.legend(title="City Type", handles=[rural, suburban, urban], loc="upper right", shadow=True, markerscale=.4)
# Incorporate a text label regarding circle size
plt.text(42,35, "Note:")
plt.text(42,33, "Circle size coorelates with driver count per city.")
plt.show()
# Save Figure
plt.savefig("Bubble_Plot.png")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
urban_fare = merged_data.loc[merged_data["type"] == "Urban", :]
urban_fare_sum = urban_fare["fare"].sum()

rural_fare = merged_data.loc[merged_data["type"] == "Rural", :]
rural_fare_sum = rural_fare["fare"].sum()

sub_fare = merged_data.loc[merged_data["type"] == "Suburban", :]
sub_fare_sum = sub_fare["fare"].sum()

# Build Pie Chart
city_type = ["Urban", "Suburban", "Rural"]
fare_sum = [urban_fare_sum, sub_fare_sum, rural_fare_sum]
colors = ["gold", "lightskyblue", "lightcoral"]
explode = [0.07, 0, 0]

plt.pie(fare_sum, explode=explode, labels=city_type, colors=colors, autopct="%1.1f%%", shadow=True, startangle=100)
plt.axis("equal")
plt.title("% of Total Fares by City Type")
plt.show()
# Save Figure
plt.savefig("Total_Fare_by_City.png")
```

## Total Rides by City Type

```python
# Calculate Ride Percents
urban_riders = merged_data.loc[merged_data["type"] == "Urban", :]
urban_riders_sum = urban_riders["ride_id"].count()

rural_riders = merged_data.loc[merged_data["type"] == "Rural", :]
rural_riders_sum = rural_riders["ride_id"].count()

sub_riders = merged_data.loc[merged_data["type"] == "Suburban", :]
sub_riders_sum = sub_riders["ride_id"].count()

# Build Pie Chart
city_type = ["Urban", "Suburban", "Rural"]
rider_sum = [urban_riders_sum, sub_riders_sum, rural_riders_sum]
colors = ["gold", "lightskyblue", "lightcoral"]
explode = [0.07, 0, 0]

plt.pie(rider_sum, explode=explode, labels=city_type, colors=colors, autopct="%1.1f%%", shadow=True, startangle=50)
plt.axis("equal")
plt.title("% of Total Riders by City Type")
plt.show()
# Save Figure
plt.savefig("Total_Rides_by_City.png")
```
